```python
import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder

import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_excel("/content/Online retail.xlsx")
```

```python
transactions = df.iloc[:, 0].apply(lambda x: x.split(','))
```

```python
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
basket_df = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
basket_df
```

```python
frequent_itemsets = apriori(basket_df, min_support=0.02, use_colnames=True)
```

```python
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric="lift",num_itemsets=7500, min_threshold=1.0)
```

```python
rules = rules.sort_values(by="lift", ascending=False)
print(rules.head(20))
```

```python
rules[rules.lift>1]
```

# **Interview Questions:**

**1.	What is lift and why is it important in Association rules?**

Lift is a measure used in association rule mining to evaluate the strength of a rule. It tells us how much more likely the items on the right-hand side (consequent) occur with the items on the left-hand side (antecedent) compared to if they were independent.

**2. What is support and Confidence. How do you calculate them?**

Support measures how frequently an itemset appears in the dataset. It tells us the proportion of transactions that contain a particular item or combination of items.

Support(A)=Number of Transactions Containing A / Total number of transactions

**3.	What are some limitations or challenges of Association rules mining?**

1. Large Number of Rules

Problem: Association rule mining can generate thousands or even millions of rules, especially in large datasets.

Impact: It becomes hard to interpret or find meaningful insights without additional filtering.

2. Requires Discrete/Transactional Data

Association rules work best on categorical or transactional data (like items bought together).

Continuous or numeric data need preprocessing or discretization, which can be time-consuming and sometimes lead to information loss.

3. Choice of Thresholds (Support & Confidence)

The results depend heavily on the minimum support and confidence thresholds.

Too high support → may miss interesting but less frequent rules.

Too low support → generates too many insignificant rules.
